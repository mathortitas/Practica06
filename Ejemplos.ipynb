# Práctica 06 - Funciones y Excepciones

## **Ejemplos**

### **1. Estructura de Excepciones**
#### **Ejemplo:**
```python
try:
    numero = int(input("Introduce un número: "))
    resultado = 10 / numero
    print("Resultado:", resultado)
except ZeroDivisionError:
    print("Error: No se puede dividir entre cero.")
except ValueError:
    print("Error: Debes introducir un número válido.")
finally:
    print("Programa finalizado.")
```

---

### **2. Estructura de Funciones**
#### **Ejemplo:**
```python
def calcular_area_circulo(radio):
    pi = 3.14159
    return pi * radio ** 2

radio = float(input("Introduce el radio del círculo: "))
print("El área del círculo es:", calcular_area_circulo(radio))
```

---

### **3. Validación de DNI**
#### **Ejemplo:**
```python
def validar_dni(dni):
    if len(dni) == 8 and dni.isdigit():
        return True
    return False

def ingresar_dni():
    while True:
        try:
            dni = input("Ingrese su DNI (8 dígitos): ")
            if not validar_dni(dni):
                raise ValueError("El DNI debe tener exactamente 8 dígitos numéricos.")
            return dni
        except ValueError as e:
            print(f"Error: {e}")

dni = ingresar_dni()
print(f"DNI ingresado correctamente: {dni}")
```

### **4. Menú de Opciones con Validación**
#### **Ejemplo:**
```python
def opcion_1():
    print("Has seleccionado la opción 1.")

def opcion_2():
    print("Has seleccionado la opción 2.")

def opcion_3():
    print("Has seleccionado la opción 3.")

def menu():
    while True:
        print("\nMenú Principal:")
        print("1. Primera Opción")
        print("2. Segunda Opción")
        print("3. Tercera Opción")
        print("4. Salir")

        try:
            opcion = int(input("Seleccione una opción: "))
            if opcion == 1:
                opcion_1()
            elif opcion == 2:
                opcion_2()
            elif opcion == 3:
                opcion_3()
            elif opcion == 4:
                print("¡Gracias por usar el programa!")
                break
            else:
                print("Opción no válida. Intente de nuevo.")
        except ValueError:
            print("Debe ingresar un número válido.")

if __name__ == "__main__":
    menu()
```
