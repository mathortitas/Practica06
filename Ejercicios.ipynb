### **1. Menú: Números Pares e Impares**
#### **Ejemplo:**
```python
def ingresar_rango():
    while True:
        try:
            inicio = int(input("Ingrese el número de inicio del rango: "))
            fin = int(input("Ingrese el número de fin del rango: "))
            if inicio > fin:
                raise ValueError("El inicio del rango debe ser menor o igual al fin.")
            return inicio, fin
        except ValueError as e:
            print(f"Error: {e}")

def es_par(numero):
    return numero % 2 == 0

def es_primo(numero):
    if numero < 2:
        return False
    for i in range(2, numero):
        if numero % i == 0:
            return False
    return True

def menu_numeros():
    rango = None
    while True:
        print("\nMenú - Números Pares e Impares:")
        print("1. Ingresar números de inicio y de fin del rango")
        print("2. Imprimir los números pares del rango")
        print("3. Imprimir los números impares del rango")
        print("4. Imprimir los números primos del rango")
        print("5. Salir")

        try:
            opcion = int(input("Seleccione una opción: "))

            if opcion == 1:
                rango = ingresar_rango()
            elif opcion == 2:
                if rango:
                    print("Números pares:")
                    for n in range(rango[0], rango[1] + 1):
                        if es_par(n):
                            print(n, end=" ")
                    print()
                else:
                    print("Primero debe ingresar un rango.")
            elif opcion == 3:
                if rango:
                    print("Números impares:")
                    for n in range(rango[0], rango[1] + 1):
                        if not es_par(n):
                            print(n, end=" ")
                    print()
                else:
                    print("Primero debe ingresar un rango.")
            elif opcion == 4:
                if rango:
                    print("Números primos:")
                    for n in range(rango[0], rango[1] + 1):
                        if es_primo(n):
                            print(n, end=" ")
                    print()
                else:
                    print("Primero debe ingresar un rango.")
            elif opcion == 5:
                print("¡Gracias por usar el programa!")
                break
            else:
                print("Opción no válida.")
        except ValueError:
            print("Debe ingresar un número válido.")

if __name__ == "__main__":
    menu_numeros()
```

### **2. Menú: Tabla de Multiplicar**
#### **Ejemplo:**
```python
def ingresar_rango():
    while True:
        try:
            inicio = int(input("Ingrese el número de inicio del rango: "))
            fin = int(input("Ingrese el número de fin del rango: "))
            if inicio > fin:
                raise ValueError("El inicio debe ser menor o igual al fin.")
            return inicio, fin
        except ValueError as e:
            print(f"Error: {e}")

def imprimir_tabla(numero):
    for i in range(1, 11):
        print(f"{numero} x {i} = {numero * i}")

def menu_tablas():
    rango = None
    while True:
        print("\nMenú - Tablas de Multiplicar:")
        print("1. Ingresar rango de números")
        print("2. Imprimir tablas de números pares")
        print("3. Imprimir tablas de números impares")
        print("4. Imprimir tabla de un número específico")
        print("5. Salir")

        try:
            opcion = int(input("Seleccione una opción: "))

            if opcion == 1:
                rango = ingresar_rango()
            elif opcion in [2, 3]:
                if rango:
                    print(f"Tablas de {'pares' if opcion == 2 else 'impares'}:")
                    for n in range(rango[0], rango[1] + 1):
                        if (opcion == 2 and n % 2 == 0) or (opcion == 3 and n % 2 != 0):
                            imprimir_tabla(n)
                else:
                    print("Primero debe ingresar un rango.")
            elif opcion == 4:
                numero = int(input("Ingrese un número: "))
                imprimir_tabla(numero)
            elif opcion == 5:
                print("¡Gracias por usar el programa!")
                break
            else:
                print("Opción no válida.")
        except ValueError:
            print("Debe ingresar un número válido.")

if __name__ == "__main__":
    menu_tablas()
```

### **3. Menú: Números Capicúa**
#### **Ejemplo:**
```python
def es_capicua(numero):
    return str(numero) == str(numero)[::-1]

def ingresar_rango_capicua():
    while True:
        try:
            inicio = int(input("Ingrese el número de inicio del rango (hasta 4 cifras): "))
            fin = int(input("Ingrese el número de fin del rango (hasta 4 cifras): "))
            if inicio > fin or inicio < 0 or fin > 9999:
                raise ValueError("Rango inválido. Use números de hasta 4 cifras.")
            return inicio, fin
        except ValueError as e:
            print(f"Error: {e}")

def menu_capicuas():
    rango = None
    while True:
        print("\nMenú - Números Capicúa:")
        print("1. Ingresar rango de números")
        print("2. Imprimir capicúas de 2 cifras")
        print("3. Imprimir capicúas de 3 cifras")
        print("4. Salir")

        try:
            opcion = int(input("Seleccione una opción: "))

            if opcion == 1:
                rango = ingresar_rango_capicua()
            elif opcion in [2, 3]:
                if rango:
                    print(f"Capicúas de {opcion + 1} cifras:")
                    for n in range(rango[0], rango[1] + 1):
                        if len(str(n)) == opcion + 1 and es_capicua(n):
                            print(n, end=" ")
                    print()
                else:
                    print("Primero debe ingresar un rango.")
            elif opcion == 4:
                print("¡Gracias por usar el programa!")
                break
            else:
                print("Opción no válida.")
        except ValueError:
            print("Debe ingresar un número válido.")

if __name__ == "__main__":
    menu_capicuas()
```

### **4. Menú: Datos del Postulante**
#### **Ejemplo:**
```python
def validar_cadena(cadena):
    return cadena.isalpha()

def ingresar_postulante():
    datos = {}
    while True:
        try:
            print("\nMenú - Datos del Postulante:")
            print("1. Nombre")
            print("2. Apellido")
            print("3. DNI")
            print("4. Celular 1")
            print("5. Celular 2")
            print("6. Altura")
            print("7. Peso")
            print("8. Imprimir datos ingresados")
            print("9. Salir")

            opcion = int(input("Seleccione una opción: "))

            if opcion == 1:
                nombre = input("Ingrese el nombre: ")
                if not validar_cadena(nombre):
                    raise ValueError("El nombre solo debe contener letras.")
                datos["Nombre"] = nombre
            elif opcion == 2:
                apellido = input("Ingrese el apellido: ")
                if not validar_cadena(apellido):
                    raise ValueError("El apellido solo debe contener letras.")
                datos["Apellido"] = apellido
            elif opcion in [3, 4, 5, 6, 7]:
                valor = input(f"Ingrese el valor para {'DNI' if opcion == 3 else 'Celular' if opcion in [4, 5] else 'Altura' if opcion == 6 else 'Peso'}: ")
                datos[f"Dato {opcion}"] = valor
            elif opcion == 8:
                print("Datos ingresados:")
                for clave, valor in datos.items():
                    print(f"{clave}: {valor}")
            elif opcion == 9:
                print("¡Gracias por usar el programa!")
                break
            else:
                print("Opción no válida.")
        except ValueError as e:
            print(f"Error: {e}")

if __name__ == "__main__":
    ingresar_postulante()
```

### **5. Menú: Datos del Alumno**

### **7. Menú: Datos del Paciente**
#### **Ejemplo:**
```python
def imprimir_dieta(peso, sexo):
    if sexo == "Mujer":
        if peso < 40:
            return "No hay dieta."
        elif peso <= 50:
            return "Dieta rica en vegetales."
        elif peso <= 60:
            return "Dieta rica en vegetales y frutas."
        elif peso <= 80:
            return "Dieta rica en vegetales, frutas y semillas."
        else:
            return "Debe hacerse cirugía."
    elif sexo == "Hombre":
        if peso < 60:
            return "No hay dieta."
        elif peso <= 75:
            return "Dieta rica en vegetales."
        elif peso <= 95:
            return "Dieta rica en vegetales y frutas."
        elif peso <= 120:
            return "Dieta rica en vegetales, frutas y semillas."
        else:
            return "Debe hacerse cirugía."

def menu_paciente():
    datos = {}
    while True:
        try:
            print("\nMenú - Datos del Paciente:")
            print("1. Nombre")
            print("2. Apellido")
            print("3. Edad")
            print("4. Peso")
            print("5. Sexo (Hombre/Mujer)")
            print("6. Imprimir Dieta")
            print("7. Salir")

            opcion = int(input("Seleccione una opción: "))

            if opcion == 1:
                datos["Nombre"] = input("Ingrese el nombre: ")
            elif opcion == 2:
                datos["Apellido"] = input("Ingrese el apellido: ")
            elif opcion == 3:
                datos["Edad"] = int(input("Ingrese la edad: "))
            elif opcion == 4:
                datos["Peso"] = float(input("Ingrese el peso en kilogramos: "))
            elif opcion == 5:
                sexo = input("Ingrese el sexo (Hombre/Mujer): ").capitalize()
                if sexo not in ["Hombre", "Mujer"]:
                    raise ValueError("Sexo no válido. Use 'Hombre' o 'Mujer'.")
                datos["Sexo"] = sexo
            elif opcion == 6:
                peso = datos.get("Peso")
                sexo = datos.get("Sexo")
                if peso is None or sexo is None:
                    print("Debe ingresar el peso y el sexo antes de imprimir la dieta.")
                else:
                    print(imprimir_dieta(peso, sexo))
            elif opcion == 7:
                print("¡Gracias por usar el programa!")
                break
            else:
                print("Opción no válida.")
        except ValueError as e:
            print(f"Error: {e}")

if __name__ == "__main__":
    menu_paciente()
```
